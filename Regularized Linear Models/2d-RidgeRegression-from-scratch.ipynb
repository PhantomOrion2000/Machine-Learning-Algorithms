```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
X, y = make_regression(n_samples = 100, n_features = 1, n_informative = 1, n_targets = 1, noise = 20, random_state =30)
```

```python
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression, Ridge
rr1 = Ridge(alpha = 10)
rr2 = Ridge(alpha = 100)
lr = LinearRegression()
```

```python
lr.fit(X,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
rr1.fit(X,y)
print(rr1.coef_)
print(rr1.intercept_)
```

```python
rr2.fit(X,y)
print(rr2.coef_)
print(rr2.intercept_)
```

```python
plt.figure(figsize = (10,6))
plt.plot(X, y, 'b+')
plt.plot(X,lr.predict(X), color = 'red', label = 'alpha = 0')
plt.plot(X,rr1.predict(X), color = 'blue', label = 'alpha = 10')
plt.plot(X,rr2.predict(X), color = 'green', label = 'alpha = 100')
plt.legend()
```
