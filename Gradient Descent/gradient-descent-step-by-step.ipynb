```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples = 4, n_features = 1,n_informative = 1, n_targets = 1, noise = 80)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
#lets apply Ordinary least square formula
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color = 'red')
```

```python
# Lets apply Gradient Descent assuming slope is constant m = 82.25
# and let's assume the starting value for intercept b = 0
y_pred = ((82.25 * X) + 0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color = 'red', label= 'OLS')
plt.plot(X, y_pred, color = 'cyan', label = 'b = 0')
plt.legend()
plt.show()
```

```python
m = 82.25
b = 100
```

```python
loss_slope = -2 * np.sum(y - m * X.ravel() - b)
loss_slope
```

```python
#lets take learning rate = 0.1
lr = 0.1

step_size = lr * loss_slope
step_size
```

```python
# Now Calculating the new intercept
b = b - step_size
b
```

```python
y_pred1 = ((82.25 * X) + b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
#Iteration 2
loss_slope = -2 * np.sum(y - m * X.ravel() - b)
loss_slope
```

```python
step_size = lr * loss_slope
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((82.25 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 3
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred3 = ((82.25 * X) + b).reshape(4)

plt.figure(figsize=(15,15))
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred3,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred2,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
b = -100
m = 82.25
lr = 0.01
epochs = 100

plt.figure(figsize=(15,15))

for i in range(epochs):
    loss_slope = -2 * np.sum(y - m*X.ravel() - b)
    b = b - (lr * loss_slope)
    
    y_pred = m*X + b
    
    plt.plot(X,y_pred)
    
plt.scatter(X,y)    
```
